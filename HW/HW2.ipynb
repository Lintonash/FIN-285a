```python
import numpy as np
import matplotlib.pyplot as plt
X = np.arange(0,10.1,0.1)
Y1 = 2*X
Y2 = (np.square(X))/5
plt.scatter(X, Y1, label = 'Y1 = 2*X' )
plt.scatter(X, Y2, label = 'Y2 = X/5' )

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

plt.plot(X, Y1, label = 'Y1 = 2*X' )
plt.plot(X, Y2, label = 'Y2 = X^2/5' )

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
from scipy.stats import norm


def cdf(mean, var, distance):
    cdf = 1- (norm.cdf(mean + distance * np.sqrt(var), mean, np.sqrt(var)) - norm.cdf(mean - distance * np.sqrt(var), mean,  np.sqrt(var)))
    return cdf

cdf1 = cdf(0,1,1.5)
cdf2 = cdf(0,1,1.5)
print(cdf1)
print(cdf2)

print('Does not depend on mean or standard deviation.')
```

```python
import pandas as pd
import numpy as np


df = pd.DataFrame([['Steve', 90, 'MF'], ['David', 85, 'MA']], columns = ['Name', 'Score', 'Major'])

df = df.set_index('Name')
print(df)

df.loc['Steve','Score'] = 92
df.loc['David', 'Major'] = 'MF'
print(df)

df.loc['John'] = [95, 'MA']
print(df)

score_mean = np.mean(df['Score'])
score_std  = np.std(df['Score'])

print(score_mean)
print(score_std)
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt


sample = [75, 150, 50, 75, 80, 25, 40, 80, 100, 10, 200, 210, 150, 80, 50, 50, 19, 100, 100, 120, 150, 170, 150, 200]
times = 100000
condition_list = []
shutdown = 1
operation = 0
i = 1
past_3_sum_list  = []
while True:
    past_3_sum = sum(np.random.choice(sample, 3))
    past_3_sum_list.append(past_3_sum)
    if past_3_sum > 400:
        condition_list.append(shutdown)
    else:
        condition_list.append(operation)
    i += 1
    if i > times:
        break

prob = sum(condition_list)/times
print(prob) 

plt.hist(past_3_sum_list,bins=20,density=True) 
(mu,sigma)= norm.fit(past_3_sum_list)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, sigma)
plt.title(title)
```
